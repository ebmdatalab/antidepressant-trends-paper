```python
import os
from pathlib import Path
import pandas as pd
from rd import get_data
```

```python
cwd = os.getcwd()
parent = str(Path(cwd).parents[0])
csv_path = os.path.join(parent,"antidepressants_last_12_months.csv")
last_12_months = pd.read_csv(csv_path, index_col="practice")
last_12_months.head()
```

```python
prescribing = last_12_months[['Dosulepin','Trimipramine']].sum(axis=1).to_frame(name='numerator')
prescribing['denominator'] = last_12_months['total_items']
prescribing['ratio'] = prescribing['numerator'] / prescribing['denominator']
prescribing.head()
```

```python
regressors = get_data()
regressors.head()
```

```python
combined = prescribing.join(regressors)
combined.to_csv('data_for_stata.csv')
combined.head()
```
